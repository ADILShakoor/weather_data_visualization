```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
path = Path('sitka_weather_2021_simple.csv')
lines = path.read_text().splitlines()
reader = csv.reader(lines)
print(reader)
header_now=next(reader)
print(header_now)

for index,colum_header in enumerate(header_now):
    print(index,colum_header)
```

```python
# poll the TMAX entrys
date,highs,lows = [],[],[]
for row in reader:
    current_date=datetime.strptime(row[2],'%Y-%m-%d')
    value = int(row[4])
    lower=int(row[5])
    date.append(current_date)
    highs.append(value)
    lows.append(lower)
# print(highs[:11])
# print(date[:11])
# print(lows[:11])
```

```python
# plot the graph for data
# plt.style.use('seaborn')
 # Plot the high temperatures.
# plt.style.use('seaborn')

fig, ax = plt.subplots()
ax.plot(date, highs, color='red',alpha=0.5)
ax.plot(date, lows, color='blue',alpha=0.1)
ax.fill_between(date, highs, lows, facecolor='blue', alpha=0.1)
 # Format plot.
ax.set_title("Daily High and Low Temperatures, 2021", fontsize=16)
ax.set_xlabel('', fontsize=10)
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=10)
ax.tick_params(labelsize=10)
plt.grid()
plt.show()
```
